```python
user={'name':'saravanan','city':'chennai','age':38}
print(user)
```

```python
print(user['city'])
```

```python
user.get('city')
```

```python
keys=user.keys()
print(keys)
```

```python
values=user.values()
print(values)
```

```python
if 'name' in user:
  print('yes')
```

```python
if 'names' in user:
  print('yes')
else:
  print('No')
```

```python
user['city']='Mumbai'
print(user)
```

```python
user.update({'mobile':9898838})
print(user)
```

```python
user.pop('mobile', None)
print(user)
```

```python
for x in user.keys():
  print(user[x])
```
